## Imports and Data

```python
import matplotlib.pyplot as plt
```

```python
epoch = 8
```

```python
lr_5e_05 = ('5e_05', [0.08048289738430583, 0.20547935773499684, 0.3216135952708843, 0.5200226244343892, 0.5200214822771213, 0.5113378684807256, 0.5672900199215989, 0.5602143887858173])
lr_3e_05 = ('3e_05', [0.07589285714285714, 0.2802099835367937, 0.5239240140500645, 0.6545634643638273, 0.6899132446500869, 0.6846849245830571, 0.6862301217335143, 0.6884976954296594])
lr_2e_05 = ('2e_05', [0.10860484544695072, 0.42453606151085144, 0.5594092182327477, 0.7194470278229869, 0.7149276383099796, 0.7079735740450027, 0.7281931836279661, 0.7142474809141477])
lr_1e_05 = ('1e_05', [0.06516290726817042, 0.24675324675324675, 0.5842891002554867, 0.6294093705858411, 0.6860166288737718, 0.7122598430869107, 0.7346648060933775, 0.7492931353690847])
lr_5e_06 = ('5e_06', [0.06896551724137931, 0.15912087912087913, 0.24770044770044766, 0.4942900237017884, 0.5530499848910312, 0.5631834281540739, 0.6011988011988011, 0.6061115355233001])
x = list(range(1, epoch + 1))
```

## Draw

```python
for lr, y in [lr_5e_05, lr_3e_05, lr_2e_05, lr_1e_05, lr_5e_06]:
    plt.plot(x, y, label=lr)
plt.legend()
plt.xlabel('epochs')
plt.ylabel('f1 score')
plt.title('Hyperparameter Search')
plt.show()
```
